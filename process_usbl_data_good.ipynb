```python
## FUNCTIONS

def stamp2sec(stamp,init):
    return (stamp - stamp[0])/np.power(10,9) # Convert to sec

def get_nav_data(robot,dataset):
    path =  dataset  + robot + "/"+robot+"_navigation.csv"
    
    # Read data
    col=['idx', 'stamp']
    stamp = pd.read_csv(path+'_nav_stamp.csv', sep=',', names=col, header=0)
    
    col=['idx', 'north', 'east', 'depth']
    ned = pd.read_csv(path+'_nav_ned.csv', sep=',', names=col, header=0)
    return pd.concat([stamp['stamp'], ned[['north', 'east', 'depth']]], axis=1)


def get_modem_raw_data(robot,dataset):
    path = 'csv/' + dataset + '/' + robot
    
    # Read data
    col=['field.header.stamp', 'field.pose.pose.position.x', 'field.pose.pose.position.y', 'field.pose.pose.position.z']
    usbl = pd.read_csv(path+'_modem_raw.csv', sep=',', usecols=col, header=0)
    usbl.columns = ['stamp', 'Nm', 'Em', 'Dm']
    usbl['stamp'] = pd.to_datetime(usbl['stamp'], unit='ns')
    usbl = usbl.set_index('stamp')
    return usbl

def get_usbllong_data(robot,dataset):
    path = 'csv/' + dataset + '/' + robot
    
    # Read data
    col=['field.header.stamp', 'field.N', 'field.E', 'field.D', 'field.rssi', 'field.integrity', 'field.accuracy']
    usbl = pd.read_csv(path+'_usbllong.csv', sep=',', usecols=col, header=0)
    usbl.columns = ['stamp', 'N', 'E', 'D', 'rssi', 'integrity', 'accuracy']
    usbl['stamp'] = pd.to_datetime(usbl['stamp'], unit='ns')
    usbl = usbl.set_index('stamp')
    return usbl

def get_channel_data(robot,dataset):
    path = 'csv/' + dataset + '/' + robot
    
    # Read data
    col=['field.header.stamp', 'field.msg_type', 'field.src_address', 'field.dst_address', 'field.duration', 'field.rssi', 'field.integrity']
    channel = pd.read_csv(path+'_channel.csv', sep=',', usecols=col, header=0)
    channel.columns = ['stamp', 'type', 'src_address', 'dst_address', 'duration', 'rssi', 'integrity']
    channel['stamp'] = pd.to_datetime(channel['stamp'], unit='ns')
    channel = channel.set_index('stamp')
    return channel
```
